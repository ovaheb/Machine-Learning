```python
j=0
h='d'
x=tr.find_all('div', {'class':'m3g-ct-row'} )
for i in x:
    y=i.find('div',{'class':'m3g-ct-col ct-team-name'})
    if(j==1):
        print(y.get_text())
        h=y.get_text() 
    j+=1
    if(y.get_text()==h):
        #print(y)
        q=i.find_all('span',{'class':'ct-player-name ct-permanent has-cat'})
        print(q[0].get_text())
```

```python
import re
j_counter=0
i_counter=0
teams=[]
player_in_definitely=[]
player_in_likely=[]
player_in_loan =[]
player_in_comebackable =[]
player_out_definitely=[]
player_out_likely=[]
player_out_loan =[]
player_out_comebackable =[]
year=[]
leage=[]
for j in (transfer.transfers):
    tr = BeautifulSoup(transfer.html[j_counter], 'html.parser')  
    j_counter+=1
    i_counter=0
    names=tr.find_all(class_='m3g-ct-col ct-team-name')
    for i in ((names)):
        if(i_counter==1):
            #print(i.get_text().split('\n')[2])
            name=i.get_text().split('\n')[2]
            #print(name)
            teams.append(name)
            flag=re.split('(\d+)',j)
            leage.append(flag[0])
            year.append(j.replace(flag[0], ''))
            x=tr.find_all('div', {'class':'m3g-ct-row'} )
            for ii in x:            
                y=ii.find('div',{'class':'m3g-ct-col ct-team-name'})
                #print(y.get_text())
                if( name in y.get_text()):
                    jj=ii.find('div',{'class':'m3g-ct-col ct-in'})
                    player1=jj.find_all('span',{'class':'ct-player-name ct-permanent has-cat'})+jj.find_all('span',{'class':'ct-player-name ct-permanent'})
                    player2=jj.find_all('span',{'class':'ct-player-name ct-permanent likely has-cat'})+jj.find_all('span',{'class':'ct-player-name ct-permanent likely'})
                    player3=jj.find_all('span',{'class':'ct-player-name ct-loan has-cat'})+jj.find_all('span',{'class':'ct-player-name ct-loan'})
                    player4=jj.find_all('span',{'class':'ct-player-name ct-loan-b'})+jj.find_all('span',{'class':'ct-player-name ct-loan-b has-cat'})
                    kk=ii.find('div',{'class':'m3g-ct-col ct-out'})
                    player5=kk.find_all('span',{'class':'ct-player-name ct-permanent has-cat'})+kk.find_all('span',{'class':'ct-player-name ct-permanent'})
                    player6=kk.find_all('span',{'class':'ct-player-name ct-permanent likely has-cat'})+kk.find_all('span',{'class':'ct-player-name ct-permanent likely'})
                    player7=kk.find_all('span',{'class':'ct-player-name ct-loan'})+kk.find_all('span',{'class':'ct-player-name ct-loan has-cat'})
                    player8=kk.find_all('span',{'class':'ct-player-name ct-loan-b'})+kk.find_all('span',{'class':'ct-player-name ct-loan-b has-cat'})
                    m=' '
                    for q in player1:
                        #print(q)
                        m=m+q.get_text().split('\n')[1] + ', '
                        #print(m)
                    player_in_definitely.append(m)
                    
                    
                    m=' '
                    for q in player2:
                        #print(q)
                        m=m+q.get_text().split('\n')[1]+ ', '
                    player_in_likely.append(m)
                    
                    
                    m=' '
                    for q in player3:
                        #print(q)
                        m=m+q.get_text().split('\n')[1]+ ', '
                    player_in_loan.append(m)
                    
                    
                    m=' '
                    for q in player4:
                        #print(q)
                        m=m+q.get_text().split('\n')[1]+ ', '
                    player_in_comebackable.append(m)
                    
                    
                    m=' '
                    for q in player5:
                        #print(q)
                        m=m+q.get_text().split('\n')[1]+ ', '
                    player_out_definitely.append(m)
                    
                    
                    m=' '
                    for q in player6:
                        #print(q)
                        m=m+q.get_text().split('\n')[1]+ ', '
                    player_out_likely.append(m)
                    
                    
                    m=' '
                    for q in player7:
                        #print(q)
                        m=m+q.get_text().split('\n')[1]+ ', '
                    player_out_loan.append(m)
                    
                    
                    m=' '
                    for q in player8:
                        #print(q)
                        m=m+q.get_text().split('\n')[1]+ ', '
                    player_out_comebackable.append(m)
                    break
        
        i_counter=1
    
    
    
    
    
    
    
    
    
    
#print(players)   
print(len(year)) 
print(len(leage)) 
d = {'لیگ':leage,'فصل':year,'تیم': teams, 'ورودی قطعی': player_in_definitely, 'ورودی احتمالی': player_in_likely, 'ورودی قرضی': player_in_loan, 'ورودی بازگشتی': player_in_comebackable, 'خروجی قطعی': player_out_definitely, 'خروجی احتمالی': player_out_likely, 'خروجی قرضی': player_out_loan, 'خروجی بازگشتی': player_out_comebackable}
df = pd.DataFrame(data=d)    
```

```python
from openpyxl import Workbook, load_workbook
from openpyxl.drawing.image import Image
from openpyxl.utils.dataframe import dataframe_to_rows
wb = load_workbook('transfers_finall.xlsx')
active = wb['Sheet1']
for x in dataframe_to_rows(df):
    active.append(x)
    
wb.save('transfers_finall.xlsx')
```
