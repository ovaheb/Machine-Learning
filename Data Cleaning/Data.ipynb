## Jame Hazfi:
Now, I loaded the csv file of jame hazfi and elaborated on its 10 first rows.

```python
Jame_hazfi = pd.read_csv('Jam-e-hazfi.csv')
Jame_hazfi.head(10)
```

Then, we inspected the html column of this dataset and tried to find important data available in it which are year, stage, team1, team2, score1, score2, time, date, penalty1, and penalty2. I tried to xtract each of them independently in this part in order to be able to explain them but in the further steps explainations are similar to here. 

```python
soup_jam = BeautifulSoup(Jame_hazfi.html[0], 'html.parser')
soup_jam
```

As an example, eliciting time colun is shown below:

```python
soup_jam.find_all(class_ = 'match-time pull-right')[0].get_text()
```

In the further parts, I neede to used re library and split strings into lists so I defined this function to convert list of strings to one string.

```python
def listToString(s):  
    result = ""  
    for i in s:  
        result += i  
    return result  
```

Now that our method of data extraction is shown, We start doing the job by extracting year. This column is available in the csv file directly and does not need any extraction.

```python
year = []
for i in range(len(Jame_hazfi)):
    year.append(Jame_hazfi.Year[i])
```

Now, we extract the stage of each match by finding its class in html and replacing the stage using some if statements.

```python
stage = []
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    temp = soup_jam.find_all(class_ = 'match-stagename col-md-3 text-left')[0].get_text().replace('جام حذفی ایران', '')
    if 'چهارم' in temp:
        stage.append('1/4')
    elif 'شانزدهم' in temp:
        stage.append('1/16')
    elif 'هشتم' in temp:
        stage.append('1/8')
    elif 'نيمه' in temp:
        stage.append('1/2')
    elif 'نیمه' in temp:
        stage.append('1/2')  
    elif 'فينال' in temp:
        stage.append('1/1')        
    else:
        stage.append(temp.replace('نهایی', ''))
```

Now, we add team1 and team2 of each match using class of them which is team-name left and team-name right.

```python
team1 = []
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    team1.append(soup_jam.find_all(class_ = 'team-name left')[0].get_text())
```

```python
team2 = []
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    team2.append(soup_jam.find_all(class_ = 'team-name right')[0].get_text())
```

The next step is to extract scores and penalty scores using re and soup libraries.

```python
score1 = []
penalty1 = []
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    temp = listToString(re.findall(r'\(.*?\)', soup_jam.find_all(class_ = 'team-score left')[0].get_text()))
    if temp == '':
        penalty1.append('0')
    else:
        penalty1.append(temp[1])
    score1.append(soup_jam.find_all(class_ = 'team-score left')[0].get_text().replace(temp, '')) 
```

```python
score2 = []
penalty2 = []
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    temp = listToString(re.findall(r'\(.*?\)', soup_jam.find_all(class_='team-score right')[0].get_text()))
    if temp == '':
        penalty2.append('0')
    else:
        penalty2.append(temp[1])
    score2.append(soup_jam.find_all(class_='team-score right')[0].get_text().replace(temp, ''))
```

Now, we wxtract date and time using their class labels.

```python
date = []
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    date.append(soup_jam.find_all(class_='match-date pull-right')[0].get_text())
```

```python
time = []
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    time.append(soup_jam.find_all(class_='match-time pull-right')[0].get_text())
```

The final step is creating dataframe and saving it in a xlsx file.

```python
data = {'Year': year, 'Stage': stage,'team1': team1, 'team2': team2, 'score1': score1, 'score2': score2, 'time': time, 'date': date, 'Penalty1': penalty1, 'Penalty2': penalty2}
Jame_hazfi_new = pd.DataFrame(data = data)
```

## Transfers:
The next data to be check is transfers.

```python
Transfers = pd.read_csv('Transfers.csv')
Transfers.head(10)
```

As always, we check the data html text.

```python
soup_transfer = BeautifulSoup(Transfers.html[0], 'html.parser')
soup_transfer
```

We extract each data of transfer windows individually and for each window we extract each teams and 8 groups of players for each one. Also year and league is added to csv file too.

```python
table_counter = 0
team_counter = 0
team = []
year = []
league = []
definite_in = []
probable_in = []
loan_in = []
loan_in_return = []
definite_out = []
probable_out = []
loan_out = []
loan_out_return = []

for j in Transfers.transfers:
    soup_transfer = BeautifulSoup(Transfers.html[table_counter], 'html.parser')  
    table_counter += 1
    team_counter = 0
    names = soup_transfer.find_all(class_ = 'm3g-ct-col ct-team-name')
    for i in names:
        if team_counter == 1:
            name = i.get_text().split('\n')[2]
            team.append(name)
            temp = re.split('(\d+)', j)
            league.append(temp[0])
            year.append(j.replace(temp[0], ''))
            temp1 = soup_transfer.find_all('div', {'class':'m3g-ct-row'})
            for cteam in temp1:            
                temp2 = cteam.find('div', {'class':'m3g-ct-col ct-team-name'})
                if name in temp2.get_text():
                    cplayer = cteam.find('div', {'class':'m3g-ct-col ct-in'})
                    player1 = cplayer.find_all('span', {'class':'ct-player-name ct-permanent has-cat'}) + cplayer.find_all('span', {'class':'ct-player-name ct-permanent'})
                    
                    player2 = cplayer.find_all('span', {'class':'ct-player-name ct-permanent likely has-cat'}) + cplayer.find_all('span', {'class':'ct-player-name ct-permanent likely'})
                    
                    player3 = cplayer.find_all('span', {'class':'ct-player-name ct-loan has-cat'}) + cplayer.find_all('span', {'class':'ct-player-name ct-loan'})
                    
                    player4 = cplayer.find_all('span', {'class':'ct-player-name ct-loan-b'}) + cplayer.find_all('span', {'class':'ct-player-name ct-loan-b has-cat'})
                    
                    cplayero = cteam.find('div', {'class':'m3g-ct-col ct-out'})
                    
                    player5 = cplayero.find_all('span', {'class':'ct-player-name ct-permanent has-cat'}) + cplayero.find_all('span', {'class':'ct-player-name ct-permanent'})
                    
                    player6 = cplayero.find_all('span', {'class':'ct-player-name ct-permanent likely has-cat'}) + cplayero.find_all('span', {'class':'ct-player-name ct-permanent likely'})
                    
                    player7 = cplayero.find_all('span', {'class':'ct-player-name ct-loan'}) + cplayero.find_all('span', {'class':'ct-player-name ct-loan has-cat'})
                    
                    player8 = cplayero.find_all('span', {'class':'ct-player-name ct-loan-b'}) + cplayero.find_all('span', {'class':'ct-player-name ct-loan-b has-cat'})
                    
                    player = ' '
                    for element in player1:
                        player += element.get_text().split('\n')[1] + ', '
                    definite_in.append(player)
                    player = ' '
                    for element in player2:
                        player +=  element.get_text().split('\n')[1] + ', '
                    probable_in.append(player)
                    player = ' '
                    for element in player3:
                        player += element.get_text().split('\n')[1] + ', '
                    loan_in.append(player)
                    player = ' '
                    for element in player4:
                        player += element.get_text().split('\n')[1] + ', '
                    loan_in_return.append(player)
                    player = ' '
                    for element in player5:
                        player += element.get_text().split('\n')[1] + ', '
                    definite_out.append(player)
                    player = ' '
                    for element in player6:
                        player += element.get_text().split('\n')[1] + ', '
                    probable_out.append(player)
                    player = ' '
                    for element in player7:
                        player += element.get_text().split('\n')[1] + ', '
                    loan_out.append(player)
                    player = ' '
                    for element in player8:
                        player += element.get_text().split('\n')[1] + ', '
                    loan_out_return.append(player)
                    break
        team_counter = 1
```

Now we save the extracted data ro a xlsx file.

```python
data = {'league':league, 'year':year, 'team': team, 'definite in': definite_in, 'probable in': probable_in, 'loan in': loan_in, 'loan in return': loan_in_return, 'definite out': definite_out, 'probable out': probable_out, 'loan out': loan_out, 'loan out return': loan_out_return}
Transfers_new = pd.DataFrame(data = data) 
```

```python
wb = Workbook()
ws =  wb.active
ws.title = "Cleaned Sheet"
wb.save(filename = 'Transfers_Cleaned.xlsx')
wb = load_workbook('Transfers_Cleaned.xlsx')
active = wb['Cleaned Sheet']
for row in dataframe_to_rows(Transfers_new):
    active.append(row)
active.delete_rows(active.min_row + 1, 1)
wb.save('Transfers_Cleaned.xlsx')
```
