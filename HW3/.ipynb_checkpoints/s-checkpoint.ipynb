```python
from sklearn import cluster, datasets, mixture
import numpy as np
import matplotlib.pyplot as plt

noisy_moons = datasets.make_moons(n_samples=500, noise=.11)

moon0 = noisy_moons[0][noisy_moons[1]==0]
moon1 = noisy_moons[0][noisy_moons[1]==1]
```

```python
class optimal_bayes:    
    def fit(self,x,y):
        self.grouping(x,y)
        self.getMeanAndCov(x.shape[0])
        
    def getMeanAndCov(self,total):
        self.means = {}
        self.covs = {}
        self.prior = {}
        for i in self.groups:
            self.means[i] = np.mean(self.groups[i],axis=0)
            self.covs[i] = np.cov(self.groups[i].T)
            self.covs[i][0][1] = self.covs[i][1][0] = 0
            self.prior[i] = len(self.groups[i])/total
        
    def grouping(self,x,y):
        self.groups = {}
        for i in range(len(x)):
            if not y[i] in self.groups:
                self.groups[y[i]] = np.asarray([x[i]])
            else:
                self.groups[y[i]] = np.vstack((x[i],self.groups[y[i]]))
            
    def probability(self, x,y): 
        mean = self.means[y]
        cov = self.covs[y]
        dim = x.shape[1]
        prob = (1/(((2*np.pi)**(dim/2))*(np.linalg.det(cov)**0.5)))
        prob *= np.exp(-0.5*np.multiply((x-mean).dot(np.linalg.inv(cov)),(x-mean))) 
        return prob.sum(axis=1)
    
    def pred(self,x):
        probs = []
        for cls in self.groups:
            prob = self.probability(x,cls)
            probs.append(prob*self.prior[cls])
        probs = np.asarray(probs)
        probs = probs.T.reshape(probs.shape[1],probs.shape[0])
        preds = np.argmax(probs,axis=1)
        classes = list(self.groups.keys())
        preds = np.asarray([classes[i] for i in preds])
        return preds
```

```python
clf0 = optimal_bayes()
clf0.fit(moon0,[0]*len(moon0))
clf1 = optimal_bayes()
clf1.fit(moon1,[1]*len(moon1))
```

```python
plt.scatter(moon0[:,0],moon0[:,1])
plt.scatter(moon1[:,0],moon1[:,1])

x = np.linspace(-2., 3.,2000)
y = np.linspace(-2., 2.,2000)
X, Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
Z = clf0.probability(XX,0)
Z = Z.reshape(X.shape)
plt.contour(X, Y, Z,levels=[x/10 for x in range(8,15)])

Z = clf1.probability(XX,1)
Z = Z.reshape(X.shape)
plt.contour(X, Y, Z,levels=[x/10 for x in range(8,15)])
```

```python
from scipy.stats import multivariate_normal
    
class GMM:
    def fit(self,x,k,itrs=400):
        N = len(x)
        f = len(x[0])
        self.means = np.asanyarray([x[int(i*N/k)] for i in range(k)])
        self.covs = [np.eye(f)]*k
        self.ps = [1/k]*k
        for _ in range(itrs):
            probs = []
            for i in range(k):
                cov = self.covs[i]+1e-6*np.identity(f)
                mean = self.means[i]
                probs.append(self.ps[i]*multivariate_normal.pdf(x,mean=mean,cov=cov))
            probs = np.asarray(probs)
            probs /= probs.sum(axis=0) 
            self.ps = probs.sum(axis=1)/N
            self.means = []
            for i in range(k):
                prob = probs[i].reshape(probs[i].shape[0],1)
                self.means.append(np.multiply(prob,x).sum(axis=0)/prob.sum())
                self.covs[i] = ((1/prob.sum())*np.dot((probs.T[:,i].reshape(N,1)*(x-self.means[-1])).T,(x-self.means[-1])))
            self.means = np.asanyarray(self.means)
        
    def show(self,figer,x):
        N = x.shape[0]
        f = x.shape[1]
        k = self.means.shape[0]
        X,Y = np.meshgrid(np.sort(x[:,0]),np.sort(x[:,1]))
        XY = np.array([X.flatten(),Y.flatten()]).T
        for i in range(k):
            probs = multivariate_normal.pdf(XY,mean=self.means[i],cov=self.covs[i]).reshape(N,N)
            figer.contour(np.sort(x[:,0]),np.sort(x[:,1]),probs,alpha=0.4)
                
    def prob(self,data):
        probs = [self.ps[0]*multivariate_normal.pdf(data,mean=self.means[0],cov=self.covs[0])]
        k = len(self.means[0])
        for i in range(k-1):
            probs.append(self.ps[i]*multivariate_normal.pdf(data,mean=self.means[i],cov=self.covs[i]))
        probs = np.asarray(probs)
        return probs.sum(axis=0)
    
    def aic(self,data):
        N = len(data)
        pred = self.prob(data)
        k = self.means.shape[0]
        ll = np.log(pred.sum())
        return (-2*ll +2*k)/N
    
    def bic(self,data):
        N = len(data)
        pred = self.prob(data)
        k = self.means.shape[0]
        ll = np.log(pred.sum())
        return -2*ll +np.log(N)*k
```

```python
plt.scatter(moon0[:,0],moon0[:,1])
plt.scatter(moon1[:,0],moon1[:,1])

gmm = GMM()
gmm.fit(moon0,3,200)
gmm.show(plt,noisy_moons[0])

gmm.fit(moon1,3,200)
gmm.show(plt,noisy_moons[0])
```

```python
fig = plt.figure(figsize=(10,10))
gmm = GMM()
aic = []
bic = []
for i in range(16):
    gmm.fit(moon0,i+1,200)
    aic.append(gmm.aic(moon0))
    bic.append(gmm.bic(moon0))
    
aic = np.asarray(aic)
bic = np.asarray(bic)
plt.figure()
plt.plot(aic)
plt.title("AIC")
plt.figure()
plt.plot(bic)
plt.title("BIC")
plt.show()
```
