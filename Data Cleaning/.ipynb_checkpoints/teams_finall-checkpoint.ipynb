```python
import pandas as pd
exel=pd.read_csv('workshop_data\Teams.csv')
exel
```

```python
from bs4 import BeautifulSoup
teams = BeautifulSoup(exel.html[33], 'html.parser')
teams
```

```python
Competition=[]
name=[]
playes=[]
score=[]
for i in range(len(exel)):
    teams = BeautifulSoup(exel.html[i], 'html.parser')  
    
    #print(i)
    x=teams.find('tr', {'class':'highlight'} )
    if(x is not None):        
        y=x.find_all('td', {'class':'xsmall-width'} )
        playes.append(y[1].get_text())
        score.append(y[2].get_text())
        Competition.append(exel.Competition[i])
        name.append(exel.team[i])
    
d = {'Competition': Competition,'name': name,'playes': playes,'score': score, }
df = pd.DataFrame(data=d)
```

```python
from openpyxl import Workbook, load_workbook
from openpyxl.drawing.image import Image
from openpyxl.utils.dataframe import dataframe_to_rows
wb = load_workbook('teams_finall.xlsx')
active = wb['Sheet1']
for x in dataframe_to_rows(df):
    active.append(x)
    
wb.save('teams_finall.xlsx')
```
