```python
import re
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1  

j=0
team1=[]
team2=[]
score1=[]
score2=[]
stage=[]
year=[]
date=[]
time=[]
penalty1=[]
penalty2=[]
df = pd.DataFrame(columns=['A'])
for i in range(len(Jame_hazfi)):
    soup_jam = BeautifulSoup(Jame_hazfi.html[i], 'html.parser')
    
    year.append(Jame_hazfi.Year[i])
    x_stage1=soup_jam.find_all(class_='match-stagename col-md-3 text-left')[0].get_text()
    x_stage=x_stage1.replace('جام حذفی ایران', '')
    
    if('چهارم' in x_stage):
        stage.append('1/4')
    elif('شانزدهم' in x_stage):
        stage.append('1/16')
    elif('هشتم' in x_stage):
        stage.append('1/8')
    elif('نيمه' in x_stage):
        stage.append('1/2')
    elif('نیمه' in x_stage):
        stage.append('1/2')  
    elif('فينال' in x_stage):
        stage.append('1/1')
    elif('فینال' in x_stage):
        stage.append('1/1')          
    else:
        stage.append(x_stage.replace('نهایی', ''))
        

    
    
    team1.append(soup_jam.find_all(class_='team-name left')[0].get_text())
    team2.append(soup_jam.find_all(class_='team-name right')[0].get_text())
    
    
    score1_help=(soup_jam.find_all(class_='team-score left')[0].get_text())
    listToStr = listToString(re.findall(r'\(.*?\)', score1_help))
    if(listToStr==''):
        penalty1.append('0')
    else:
        penalty1.append(listToStr[1])
    score1.append(score1_help.replace(listToStr, ''))
    
    
    
    score2_help=(soup_jam.find_all(class_='team-score right')[0].get_text())
    listToStr = listToString(re.findall(r'\(.*?\)', score2_help))
    if(listToStr==''):
        penalty2.append('0')
    else:
        penalty2.append(listToStr[1])
    score2.append(score2_help.replace(listToStr, ''))
    

    
    
    
    
    date.append(soup_jam.find_all(class_='match-date pull-right')[0].get_text())
    time.append(soup_jam.find_all(class_='match-time pull-right')[0].get_text())
    
d = {'Year': year, 'stage': stage,'team1': team1, 'team2': team2, 'score1': score1, 'score2': score2, 'time': time, 'date': date, 'penalty1': penalty1, 'penalty2': penalty2}
df = pd.DataFrame(data=d)
```

```python
from openpyxl import Workbook, load_workbook
from openpyxl.drawing.image import Image
from openpyxl.utils.dataframe import dataframe_to_rows
wb = load_workbook('WorldCup_finall.xlsx')
active = wb['Sheet1']
for x in dataframe_to_rows(df):
    active.append(x)
    
wb.save('WorldCup_finall.xlsx')
```
