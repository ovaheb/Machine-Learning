```python
import pandas as pd
import matplotlib.pyplot as plt

DataPath = 'Iris.csv'
Data = pd.read_csv(DataPath)
x = Data[Data.columns[0:4]] 
y = Data[Data.columns[4]]
colors = {'Iris-setosa':'red','Iris-virginica':'green','Iris-versicolor':'blue'}
fig, axs = plt.subplots(3,2,figsize=(12,18))

axs[0,0].scatter(x[x.columns[0]],x[x.columns[1]],c=y.map(colors))
axs[0,0].set_xlabel(x.columns[0])
axs[0,0].set_ylabel(x.columns[1])

axs[0,1].scatter(x[x.columns[0]],x[x.columns[2]],c=y.map(colors))
axs[0,1].set_xlabel(x.columns[0])
axs[0,1].set_ylabel(x.columns[2])

axs[1,0].scatter(x[x.columns[0]],x[x.columns[3]],c=y.map(colors))
axs[1,0].set_xlabel(x.columns[0])
axs[1,0].set_ylabel(x.columns[3])

axs[1,1].scatter(x[x.columns[1]],x[x.columns[2]],c=y.map(colors))
axs[1,1].set_xlabel(x.columns[1])
axs[1,1].set_ylabel(x.columns[2])

axs[2,0].scatter(x[x.columns[1]],x[x.columns[3]],c=y.map(colors))
axs[2,0].set_xlabel(x.columns[1])
axs[2,0].set_ylabel(x.columns[3])

axs[2,1].scatter(x[x.columns[2]],x[x.columns[3]],c=y.map(colors))
axs[2,1].set_xlabel(x.columns[2])
axs[2,1].set_ylabel(x.columns[3])
plt.show()


from sklearn.metrics import  confusion_matrix
import itertools
import numpy as np

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.figure(figsize = (5,5))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

def report(y_test,y_pred,title):
    print(title)
    conf_mtx = confusion_matrix(y_test,y_pred)
    FP = conf_mtx.sum(axis=0) - np.diag(conf_mtx)  
    FN = conf_mtx.sum(axis=1) - np.diag(conf_mtx)
    TP = np.diag(conf_mtx)
    TN = conf_mtx.sum() - (FP + FN + TP)
    ACC = (TP+TN)/(TP+FP+FN+TN)
    SENSITIVITY = TP/(TP+FN)
    SPECCIFITY = TN/(TN+FP) 
    PRECISION = TP/(TP+FP)
    f1_score = TP/(TP+0.5*(FP+FN))
    print("accuracy :",ACC)
    print("sensitivity:" ,SENSITIVITY)
    print("specifity:",SPECCIFITY)
    print("precision:",PRECISION)
    print("F1_score:",f1_score)
    classes = list(dict.fromkeys(y_test))
    plot_confusion_matrix(conf_mtx,classes,title=title)


Data_shuffled = Data.sample(frac=1).reset_index(drop=True)
x_shuffled = Data_shuffled[Data.columns[2:4]] 
y_shuffled = Data_shuffled[Data.columns[4]]
train_size=0.8
spliter = int(len(y)*train_size)
x_train = x_shuffled[:spliter]
y_train = y_shuffled[:spliter]
x_test = x_shuffled[spliter:]
y_test = y_shuffled[spliter:]


from math import sqrt

class Nearest_classifier:
    def fit(self,x_train,y_train):
        self.x = x_train
        self.y = y_train
        
    def predict(self,x_test):
        preds = []
        for index ,x in x_test.iterrows():
            dists = [self.calc_dist(x,xi) for index , xi in self.x.iterrows()]
            pred = self.y[dists.index(min(dists))]
            preds.append(pred)
        return preds
    
    def calc_dist(self,x1,x2):
        D = x1-x2
        D = sqrt(sum([d*d for d in D]))
        return D
    
def my_normalize(x):
    normalized = {}
    for i in x.columns:
        col = x[i]
        minn = min(col)
        maxx = max(col)
        norm_col = [(j-minn)/(maxx-minn) for j in col]
        normalized[i] = norm_col
    return pd.DataFrame(normalized)


#unnormalized
clf = Nearest_classifier()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
report(y_test,y_pred,"unnormalized_ no libarary")

#normalized
clf = Nearest_classifier()
x_train_norm = my_normalize(x_train)
x_test_norm = my_normalize(x_test)
clf.fit(x_train_norm,y_train)
y_pred = clf.predict(x_test_norm)
report(y_test,y_pred,"normalized_ no libarary")

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import plot_roc_curve,roc_curve,auc, roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(x[x.columns[2:]], y, test_size=0.2, random_state=10)

#unnormalized
clf = KNeighborsClassifier(n_neighbors=1)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
report(y_test,y_pred,"unnormalized_ using libarary")

dic = {'Iris-setosa':0,'Iris-virginica':1,'Iris-versicolor':2}
fig, axs = plt.subplots(3,figsize=(12,18))
for i in dic.keys():
    y_test_temp = [dic[j]==dic[i] for j in y_test]
    y_pred_temp = [dic[j]==dic[i] for j in y_pred]
    fpr, tpr, threshold = roc_curve(y_test_temp,y_pred_temp)
    roc_auc = auc(fpr, tpr)
    axs[dic[i]].plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    axs[dic[i]].set_title(i)
    
    print(i+':',roc_auc)
plt.show()

from sklearn.preprocessing import normalize
#normilized
X_train_normalized = normalize(X_train)
clf = KNeighborsClassifier(3)
clf.fit(X_train_normalized,y_train)
X_test_normalized = normalize(X_test)
y_pred = clf.predict(X_test_normalized)
report(y_test,y_pred,'normalized_using library')

dic = {'Iris-setosa':0,'Iris-virginica':1,'Iris-versicolor':2}
fig, axs = plt.subplots(3,figsize=(12,18))
for i in dic.keys():
    y_test_temp = [dic[j]==dic[i] for j in y_test]
    y_pred_temp = [dic[j]==dic[i] for j in y_pred]
    fpr, tpr, threshold = roc_curve(y_test_temp,y_pred_temp)
    roc_auc = auc(fpr, tpr)
    axs[dic[i]].plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    axs[dic[i]].set_title(i)
    
    print(i+':',roc_auc)
plt.show()
```
